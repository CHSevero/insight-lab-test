```python
import pandas as pd
import pandas as np
```

```python
data_1 = pd.read_csv("DadosBO_2019_1(ROUBO DE VEÍCULOS).csv")
data_2 = pd.read_csv("DadosBO_2019_2(ROUBO DE VEÍCULOS).csv")
data_3 = pd.read_csv("DadosBO_2019_3(ROUBO DE VEÍCULOS).csv")
data_4 = pd.read_csv("DadosBO_2019_4(ROUBO DE VEÍCULOS).csv")
data_5 = pd.read_csv("DadosBO_2019_5(ROUBO DE VEÍCULOS).csv")
data_6 = pd.read_csv("DadosBO_2019_6(ROUBO DE VEÍCULOS).csv")
data_7 = pd.read_csv("DadosBO_2019_7(ROUBO DE VEÍCULOS).csv")
data_8 = pd.read_csv("DadosBO_2019_8(ROUBO DE VEÍCULOS).csv")
data_9 = pd.read_csv("DadosBO_2019_9(ROUBO DE VEÍCULOS).csv")
data_10 = pd.read_csv("DadosBO_2019_10(ROUBO DE VEÍCULOS).csv")
data_11 = pd.read_csv("DadosBO_2019_11(ROUBO DE VEÍCULOS).csv")
data_12 = pd.read_csv("DadosBO_2019_12(ROUBO DE VEÍCULOS).csv")
```

```python
data = pd.concat([
    data_1,
    data_2,
    data_3,
    data_4,
    data_5,
    data_6,
    data_7,
    data_8,
    data_9,
    data_10,
    data_11,
    data_12
])
```

```python
data.columns
```

```python
# hora_dia, dia_semana, bairro e cidade
features = ['HORAOCORRENCIA', 'DATAOCORRENCIA', 'BAIRRO','CIDADE']
data = data[features]
```

```python
data.head()
```

```python
data.info()
```

```python
# TRANSFORMANDO DADOS TEMPORAIS NO TIPO APROPRIADO
data['DATAOCORRENCIA'] = pd.to_datetime(data['DATAOCORRENCIA'], format='%d/%m/%Y', errors = 'coerce')
data.head()
```

```python
data = data.dropna()
```

```python
DiaDaSemana = ['Segunda','Terça','Quarta','Quinta','Sexta','Sabado','Domingo']
DS = lambda x: DiaDaSemana[x.dayofweek]
#DS = lambda x: int(x.dayofweek)
data['DIASEMANA'] =data['DATAOCORRENCIA'].apply(DS)
data.head()
```

```python
data.DIASEMANA.unique()
```

```python
data.HORAOCORRENCIA.iloc[0].split(':')[0]
```

```python
H = lambda x: int(x.split(':')[0])
data['HORA'] =data['HORAOCORRENCIA'].apply(H)
data.head()
```

```python
data.HORA.unique()
```

```python
data.CIDADE.value_counts()
```

```python
data.DATAOCORRENCIA.iloc[0].date()
```

```python
A= lambda x: x.date()
data['DATA'] = data.DATAOCORRENCIA.apply(A)
```

```python
data.head()
```

```python
a = list(zip(data.BAIRRO,data.CIDADE,data.DIASEMANA,data.HORA,data.DATA))
```

```python
b = list()
for i in a:
    b.append(data[(data.BAIRRO==i[0])&(data.CIDADE==i[1])&(data.DIASEMANA==i[2])&(data.HORA==i[3])&(data.DATA==i[4])].count())
```
